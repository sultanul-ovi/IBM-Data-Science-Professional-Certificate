####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<details><summary>Click here for a hint</summary>

<p>
    You will require:-
    <br>create a separate dataframe where the column recession has a value of '1'
    <br>to group the year, vehicle_type and calculate the average on the 'Automobile Sales'
    <br>one way is to -
    <br>use as_index as false else you will endup with multiple-indexed datafame
    <br>later set year as index and groupby vehicle over Sales and plot
    <br>make use of .plot() with kind = 'line'
    <br>do not forget to include labels and title
</p>
</details>


````python
# Filter data for recession periods
df_rec = df[df['Recession'] == 1]

# Calculate the average automobile sales by year and vehicle type during the recession
df_Mline = df_rec.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Calculate normalized sales by dividing by the average sales for each vehicle type
df_Mline['Normalized_Sales'] = df_Mline.groupby('Vehicle_Type')['Automobile_Sales'].transform(lambda x: x / x.mean())

# Set 'Year' as the index
df_Mline.set_index('Year', inplace=True)

# Create the line plot for each vehicle type
plt.figure(figsize=(12, 8))
for vehicle_type in df_Mline['Vehicle_Type'].unique():
    data = df_Mline[df_Mline['Vehicle_Type'] == vehicle_type]
    plt.plot(data.index, data['Normalized_Sales'], label=vehicle_type, marker='o')

# Highlight recession years
recession_years = df_rec['Year'].unique()
for year in recession_years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)

# Add labels, legend, and title
plt.legend(title="Vehicle Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel("Normalized Sales")
plt.xlabel("Year")
plt.title("Normalized Automobile Sales by Vehicle Type During Recession")

# Save and display the plot
plt.savefig("Line_plot_2.png")  # Saves the image
plt.show()
````

### Now you want to compare the sales of different vehicle types during a recession and a non-recession period
<br>We recommend that you use the functionality of **Seaborn Library** to create this visualization


<details><summary>Click here for a hint</summary>

<p>
     To visualize sales of different vehicles during recession and non-recession periods, you can use a bar chart
        <br> You will need to group Recession, Vehicle_Type for average Automobile_Sales and then plot it<br>

````python
import numpy as np
# Group the data by Recession and Vehicle_Type and calculate the mean Automobile_Sales
dd = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Pivot the data to get a better structure for grouped bar plotting
pivot_df = dd.pivot(index='Vehicle_Type', columns='Recession', values='Automobile_Sales')
pivot_df = pivot_df.fillna(0)  # Fill NaN values with 0 if any

# Set up the figure and bar width
bar_width = 0.35
index = np.arange(len(pivot_df))

# Plot the bars for Non-Recession and Recession
fig, ax = plt.subplots(figsize=(12, 6))
bar1 = ax.bar(index, pivot_df[0], bar_width, label='Non-Recession', color='blue')
bar2 = ax.bar(index + bar_width, pivot_df[1], bar_width, label='Recession', color='orange')

# Add labels and title
ax.set_xlabel('Vehicle Type')
ax.set_ylabel('Average Automobile Sales')
ax.set_title('Vehicle-Wise Sales during Recession and Non-Recession Period')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(pivot_df.index, rotation=45)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import matplotlib.pyplot as plt

# Filter data for non-recession years
non_rec_data = df[df['Recession'] == 0]

# Define size for bubbles using Seasonality_Weight
size = non_rec_data['Seasonality_Weight'] * 100  # Scale size for better visibility

# Create bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(non_rec_data['Month'], non_rec_data['Automobile_Sales'], s=size, alpha=0.5, c='blue')

# Labels and title
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality Impact on Automobile Sales')

# Show plot
plt.show()
````
